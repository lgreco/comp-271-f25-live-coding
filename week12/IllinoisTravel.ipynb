```python
def sssp(graph, source):
    """Find the shortest path to each vertex in graph from
    a source vertex."""

    # Shortcut to number of vertices in the graph
    n = len(graph)
    NO_EDGE = graph[0][0]

    # Initialize the data structures for this algorithm
    predecessor = [ None ] * n  # How to get to every vertex in a shortest path
    distance = [ float('inf') ] * n  # the shortest path distance to each vertex
    distance[source] = 0  # Distance from source vertex to itself, is 0

    # Places to visit next
    places_to_visit_next = [ source ]

    # Explore all connections, while there are places to visit
    while len(places_to_visit_next) > 0:

        # graph some place to visit
        u = places_to_visit_next.pop(0)

        # Find all the neighbors of vertex (city) u:
        for v in range(n):
            if graph[u][v] != NO_EDGE:
                alternative_distance_to_v = distance[u] + graph[u][v]
                current_distance_to_v = distance[v]  # along the current shortest path    
                # Check if the edge from u to v is tense
                if alternative_distance_to_v < current_distance_to_v:
                    # yup, we have a tense edge, relax it
                    distance[v] = alternative_distance_to_v
                    predecessor[v] = u
                    places_to_visit_next.append(v)
    
    # We are done with all the places to explore:
    return distance, predecessor

```

#### Fibonacci's numbers

$$
F(n) = F(n-1) + F(n-2)
$$
with $n\in\mathbb{N_0}$, $F(0) = 0$, and $F(1) =1$.


```python
def fibo(n:int) -> int:
    if n < 2:
        return n
    else:
        return fibo(n-1) + fibo(n-2)
```

```python
for i in range(20):
    print(f"{i:3d} {fibo(i):20d}")
```

```python

def fibo_smart(n, fibo_list):
    fibn = n
    if n > 1:
        if n < len(fibo_list):
            fibn = fibo_list[n-1] + fibo_list[n-2]
        else:
            fibn = fibo_smart(n-1,fibo_list)+fibo_smart(n-2, fibo_list)
            fibo_list.append(fibn)
    return fibn

def fibo_api(n):
    return fibo_smart(n, [0,1])
```

```python
def super_fibo(n):
    if n < 2:
        return n
    else:
        previous = 1
        previous_previous = 0
        for i in range(1,n):
            next = previous + previous_previous
            previous_previous = previous
            previous = next
        return next
```

```python
print(time.time_ns())
print(super_fibo(5000))
print(time.time_ns())
```
